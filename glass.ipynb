```python
import numpy as np
import matplotlib.pyplot as plt

import glass as glass
```

# 透明ガラスまたは吸熱ガラスの直達日射に対する透過率、反射率を求める

```python
cos_i = np.arange(start=0, stop=1.01, step=0.01, dtype=float)
```

```python
# 透明ガラスの直達日射透過率、吸収率、反射率の計算
trans_clear, reflect_clear = glass.glass(K=0, D=0.006, CI=cos_i)

absorp_clear = 1.0 - trans_clear - reflect_clear
```

```python
# 吸収ガラスの直達日射透過率、吸収率、反射率の計算
trans_absorbing, reflect_absorbing = glass.glass(K=1, D=0.006, CI=cos_i)

absorp_absorbing = 1.0 - trans_absorbing - reflect_absorbing
```

## グラフの描画

```python
# グラフの作成
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# 透過率のグラフ
axes[0].plot(cos_i, trans_clear, label='clear')
axes[0].plot(cos_i, trans_absorbing, label='absorbing')
axes[0].set_title('transmission')
axes[0].set_xlabel('cos_i')
axes[0].set_ylabel('transmission')
axes[0].set_ylim(0, 1)
axes[0].legend()

# 吸収率のグラフ
axes[1].plot(cos_i, absorp_clear, label='clear')
axes[1].plot(cos_i, absorp_absorbing, label='absorbing')
axes[1].set_title('absorption')
axes[1].set_xlabel('cos_i')
axes[1].set_ylabel('absorption')
axes[1].set_ylim(0, 1)
axes[1].legend()

# 反射率のグラフ
axes[2].plot(cos_i, reflect_clear, label='clear')
axes[2].plot(cos_i, reflect_absorbing, label='absorbing')
axes[2].set_title('reflectivity')
axes[2].set_xlabel('cos_i')
axes[2].set_ylabel('reflectivity')
axes[2].set_ylim(0, 1)
axes[2].legend()

# グラフの表示
plt.tight_layout()
plt.show()
```
