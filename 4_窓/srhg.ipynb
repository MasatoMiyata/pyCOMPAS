```python
import numpy as np
import matplotlib.pyplot as plt

import srhg
```

# 日射熱取得量を計算する

```python
cos_i = np.arange(start=0, stop=1.01, step=0.01, dtype=float)
```

```python
# 透明ガラスまたは吸熱ガラスの直達日射透過率の計算
trans_clear, _ = srhg.srhg(itype=1, srd=1, srs=1, ci=cos_i)
```

```python
# 熱反射ガラスの直達日射透過率、吸収率、反射率の計算
trans_heat_ref, _ = srhg.srhg(itype=2, srd=1, srs=1, ci=cos_i)
```

## グラフの描画

```python
# 透過率のグラフ
plt.plot(cos_i, trans_clear, label='clear')
plt.plot(cos_i, trans_heat_ref, label='heat-reflective')
plt.suptitle('transmission')
plt.xlabel('cos_i')
plt.ylabel('transmission')
plt.ylim(0, 1)
plt.legend()

# グラフの表示
plt.tight_layout()
plt.show()
```
