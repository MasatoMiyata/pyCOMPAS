```python
import numpy as np
import matplotlib.pyplot as plt

import glasrl as gr
```

# 3mm普通ガラスの日射および昼光に対する透過率、吸収率（光の場合は反射率）を返す

```python
cos_i = np.arange(start=0, stop=1.01, step=0.01, dtype=float)
```

```python
# 直達日射透過率、吸収率、反射率の計算
trans_solar_drctrad, absorp_drctrad, reflect_drctrad = gr.glasrl(ic=1, ca=cos_i)
# 直射日光の透過率、吸収率、反射率の計算
trans_solar_drctsunlgt, absorp_drctsunlgt, reflect_drctsunlgt = gr.glasrl(ic=2, ca=cos_i)
# 拡散日射透過率、吸収率、反射率の計算
trans_solar_skyrad, absorp_skyrad, reflect_skyrad = gr.glasrl(ic=-1, ca=cos_i)
# 天空光の透過率、吸収率、反射率の計算
trans_solar_skysunlgt, absorp_skysunlgt, reflect_skysunlgt = gr.glasrl(ic=-2, ca=cos_i)
```

## グラフの描画

```python
# グラフの作成
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# 透過率のグラフ
axes[0].plot(cos_i, trans_solar_drctrad, label='direct_solar_radiation')
axes[0].plot(cos_i, trans_solar_drctsunlgt, label='direct_sunlight')
axes[0].plot(cos_i, np.full_like(cos_i, trans_solar_skyrad), label='sky_solar_radiation')
axes[0].plot(cos_i, np.full_like(cos_i, trans_solar_skysunlgt), label='sky_sunlight')
axes[0].set_title('transmission')
axes[0].set_xlabel('cos_i')
axes[0].set_ylabel('値')
axes[0].legend()

# 吸収率のグラフ
axes[1].plot(cos_i, absorp_drctrad, label='direct_solar_radiation')
axes[1].plot(cos_i, absorp_drctsunlgt, label='direct_sunlight')
axes[1].plot(cos_i, np.full_like(cos_i, absorp_skyrad), label='sky_solar_radiation')
axes[1].plot(cos_i, np.full_like(cos_i, absorp_skysunlgt), label='sky_sunlight')
axes[1].set_title('absorption')
axes[1].set_xlabel('cos_i')
axes[1].set_ylabel('値')
axes[1].legend()

# 反射率のグラフ
axes[2].plot(cos_i, reflect_drctrad, label='direct_solar_radiation')
axes[2].plot(cos_i, reflect_drctsunlgt, label='direct_sunlight')
axes[2].plot(cos_i, np.full_like(cos_i, reflect_skyrad), label='sky_solar_radiation')
axes[2].plot(cos_i, np.full_like(cos_i, reflect_skysunlgt), label='sky_sunlight')
axes[2].set_title('reflectivity')
axes[2].set_xlabel('cos_i')
axes[2].set_ylabel('値')
axes[2].legend()

# グラフの表示
plt.tight_layout()
plt.show()
```
